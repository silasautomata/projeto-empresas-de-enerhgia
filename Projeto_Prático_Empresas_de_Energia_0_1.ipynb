```python
# Bibliotecas
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
base_dados =pd.read_excel('/content/Dados_Empresas_Energia.xlsx')
base_dados.head()
```

```python
# define a coluna data com um indice- a mudança e feita no dataframe diretamente
base_dados.set_index('Data', inplace=True)

base_dados.head()
```

```python
# grafico

# estilodo grafico
#plt.style.use('seaborn-darkgrid')
# tamanho
plt.figure(figsize=(16, 7))
# title
plt.title('analise açoes de empresas de energia', loc='left',  fontsize=18, fontweight=0)

# plot da petrobas
 # pega a coluna data definida como index coma coluna petrobras, colocando a cor verde com pouca transparencia
plt.plot(base_dados.index, base_dados['Petrobras'], color='#008c4a', linewidth=4, alpha=0.7)

# texto petrobras
 # deixa o index data como o eixo X, e o ultimo valor da coluna petrobras-printa o nome petroobras com a cor verde larga- alinha ao lado esquerso
plt.text(base_dados.index[-1], base_dados['Petrobras'].tail(1), 'Petrobras', color='#008c4a', size='large', horizontalalignment='left')

# plot de todas as colunas
# pega todas as colunas do dataframe apartir da primeiro posiçao endiante
for coluna in base_dados.columns[1:]: # repete o for com todas as colunas do dataframe
  # texto das outras açoes
  plt.plot(base_dados.index, base_dados[coluna], color='gray', linewidth=2, alpha=0.7) # plot das açoes
  plt.text(base_dados.index[-1], base_dados[coluna].tail(1), coluna, color='gray', size='large', horizontalalignment='left')# texto das outras açoes

# labels
plt.xlabel('Periodo')
plt.ylabel('Preço de fechamneto (R$)');
```
